```python
team={"Chennai Super Kings":"CSK",
      "Deccan Chargers":"DC",
      "Delhi Daredevils":"DD",
      "Kolkata Knight Riders":"KKR",
      "Royal Challengers Bangalore":"RCB",
      "Gujarat Lions":"GL",
      "Kings XI Punjab":"KXIP",
      "Mumbai Indians":"MI",
      "Kochi Tuskers Kerala":"KTK",
      "Pune Warriors":"PWI",
      "Rajasthan Royals":"RR",
      "Rising Pune Supergiants":"RPS",
      "Rising Pune Supergiant":"RPS",
      "Sunrisers Hyderabad":"SRH"
      }
```

```python
data
```

```python
for i in range(len(data)):
  if(pd.isnull(data['bat_team'][i])):
    data['bat_team'][i]="NAN_bat"
  else:
    data['bat_team'][i]=team[data['bat_team'][i]]+"_bat"
  if(pd.isnull(data['bowl_team'][i])):
    data['bowl_team'][i]="NAN_bowl"
  else:
    data['bowl_team'][i]=team[data['bowl_team'][i]]+"_bowl"
  if(pd.isnull(data['venue'][i])):
    data['venue'][i]="NAN_venue"
```

```python
bat_dummies=pd.get_dummies(data['bat_team'])
bowl_dummies=pd.get_dummies(data['bowl_team'])
venue_dummies=pd.get_dummies(data['venue'])
new_data=pd.concat([bat_dummies,bowl_dummies],axis=1)
new_data=pd.concat([new_data,venue_dummies],axis=1)
```

```python
new_data
```

```python
new_data=pd.concat([new_data,data['total']],axis=1)
```

```python
new_data
```

```python
X=new_data.iloc[:,:-1].values
Y=new_data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
import random
```

```python
depth = random.randrange(2,10)
regr = RandomForestRegressor(max_depth=depth, random_state=0)
```

```python
regr.fit(x_train, y_train)
```

```python
y_pred=regr.predict(x_test)
y_pred
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
